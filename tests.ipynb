```python
# 1day interval

def get_historical_data(symbol, days_back=30):
    # Calculer la date de début comme il y a `days_back` jours à partir d'aujourd'hui
    start_date = (datetime.now() - timedelta(days=days_back)).strftime('%Y-%m-%d')

    historical_url = f'https://cloud.iexapis.com/stable/stock/{symbol}/chart/{days_back}d?token={IEX_CLOUD_API_TOKEN}'
    response = requests.get(historical_url)

    if response.status_code == 200:
        historical_data = response.json()
        # historical_data = [entry['date'] for entry in historical_data]
        return historical_data
    else:
        print(f"Erreur lors de la requête historique : {response.status_code}")
        return None


get_historical_data('AAPL')
```

```python
import yfinance as yf
import time

def get_stock_data(ticker):
    stock = yf.Ticker(ticker)
    data = stock.history(period='1y')  # Récupérer les données sur une période d'un an
    return data.to_json()

stock_data = get_stock_data('^GSPC')
print(stock_data)
```

```python
import yfinance as yf
import pandas as pd

# Obtenez les symboles des actions du S&P 500
sp500_symbols = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].tolist()

# Récupérez les données pour chaque action
stock_data = {}
for symbol in sp500_symbols:
    try:
        stock_data[symbol] = yf.download(symbol, start="2022-01-01", end="2022-12-31")
    except Exception as e:
        print(f"Erreur lors de la récupération des données pour {symbol}: {e}")

# stock_data contient maintenant les données pour chaque action
```
